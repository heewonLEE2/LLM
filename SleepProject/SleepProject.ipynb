```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv("data/Sleep_health_and_lifestyle_dataset.csv")

# 데이터 상위 5개 행 출력
df[['Heart Rate', 'Sleep Duration']].head(15)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 폰트 경로 지정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()

# 설정 강제 적용
plt.rcParams['font.family'] = font_name
plt.rcParams['axes.unicode_minus'] = False

# Seaborn 스타일
sns.set(style="whitegrid")

# 적용 확인
print("✅ 적용된 폰트 이름:", font_name)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Sleep Disorder", y="Physical Activity Level")
plt.title("수면장애에 따른 신체 활동 수준", fontproperties=font_prop)
plt.xlabel("수면장애 유무", fontproperties=font_prop)
plt.ylabel("신체 활동 수준", fontproperties=font_prop)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Sleep Disorder", y="Daily Steps")
plt.title("수면장애에 따른 하루 걸음 수", fontproperties=font_prop)
plt.xlabel("수면장애", fontproperties=font_prop)
plt.ylabel("걸음 수", fontproperties=font_prop)
plt.show()
```

```python
age_stress = df.groupby("Age")["Stress Level"].mean()

plt.figure(figsize=(10, 5))
plt.plot(age_stress.index, age_stress.values, marker='o')
plt.title("나이대에 따른 평균 스트레스 수준", fontproperties=font_prop)
plt.xlabel("나이", fontproperties=font_prop)
plt.ylabel("스트레스 수준", fontproperties=font_prop)
plt.grid(True)
plt.show()
```
